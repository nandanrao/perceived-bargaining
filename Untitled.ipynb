```
from abce import Agent, Simulation
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 7.5]
```

```
from random import randrange

class Worker(Agent):
    def init(self, parameters):
        self.num_firms = parameters['num_firms']
        self.num_workers = parameters['num_workers']
        # self.apps_per_day = agent_parameters['initial_apps_per_day']
        self.apps_per_day = 20
        self.value = (1/self.num_workers) * self.id + .001
        self.create('application', 999999999999)
        self.create('labor', 1)
        self.apps_sent = 0
        self.offers_received = 0

    def send_application(self):
        apps = self.apps_per_day
        if self['labor'] > 0:
            if self.apps_sent == 0:
                thresh = self.apps_per_day/50
            else:
                thresh = (self.apps_per_day + self.apps_sent)/50


            # direct app to only firms in your range...
            max_firm = int(self.value * self.num_firms)
            max_firm = max(max_firm, 5)
            max_firm = min(max_firm, self.num_firms)
            min_firm = int((self.value-thresh) * self.num_firms)
            min_firm = max(min_firm, 0)

            # print('worker {} has min, max: {}, {}'.format(self.id, min_firm, max_firm))

            for i in range(apps):
                firm = randrange(min_firm, max_firm)
                self.sell(('firm', firm), good = 'application', quantity = 1, price = 1/self.value)
            self.apps_sent += apps

    
    def review_offers(self):
        offers = self.get_offers('labor', descending = True)
        self.offers_received += len(offers)
        try: 
            # only accept if above threshold 
            # threshold is determined by number of offers recieved each round
            best_offer, *others = offers
            # if best_offer.price > thresh:
            self.accept(best_offer)
            # else:
                # self.reject(best_offer)
            for o in others:
                self.reject(o)
        except ValueError:
            return        

    def print_possessions(self):
        print('worker {}: {}. Ratio: {}'.format(self.id, self.possessions().get('money'), self.offers_received/self.apps_sent))

    def get_value(self):
        return self.value
    
class Firm(Agent):
    def init(self, parameters):
        val = 1 / parameters['num_firms']
        moneys = val * (self.id - 1)
        moneys = moneys if moneys > 0 else val
        self.create('money', moneys)

    def send_offers(self):
        poss = self.possessions()
        applicants = self.get_offers('application')
        if not poss.get('labor'):
            if applicants:
                best_applicant, *others = applicants
                if 1/best_applicant.price >= self['money']:
                    self.buy(best_applicant.sender, 
                             good = 'labor', 
                             quantity = 1, 
                             price = self['money'])


    def print_possesions(self):
        print('firm {}: {}'.format(self.id, self.possessions()))
    

sim = Simulation(name='foo', processes = 1)

ps = {'num_firms': 3000, 'num_workers': 3000}
firms = sim.build_agents(Firm, 'firm', number = ps['num_firms'], parameters = ps)
workers = sim.build_agents(Worker, 'worker', number = ps['num_workers'],  parameters = ps)

u = 1000
diff = 1000
# while diff > 1:
for i in range(5):
    sim.advance_round(i)
    workers.send_application()
    firms.send_offers()
    workers.review_offers()
    new_u = np.array([w.get('labor') for w in workers.possessions()]).sum()
    diff = u - new_u
    u = new_u
    print(u)
```
