```
from abce import Agent, Simulation
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [7.5, 5]
```

```
from random import randrange
from scipy import stats
import math

def get_clips(mean, std, lo, hi):
    a, b = (lo - mean) / std, (hi - mean) / std
    return a,b

class Worker(Agent):
    def init(self, parameters):
        self.num_firms = parameters['num_firms']
        self.num_workers = parameters['num_workers']
        self.apps_per_day = parameters['apps_per_day']
        self.value = (1/self.num_workers) * self.id + .001
        self.create('application', 999999999999)
        self.create('labor', 1)
        self.apps_sent = 0
        self.offers_received = 0

    def get_thresh(self):
        if self.apps_sent == 0:
            thresh = self.value - 0.05
        elif self.offers_received == 0:
            thresh = 0.0
        else:
            thresh = self.value - self.apps_sent/(self.offers_received*100)
        return thresh

    def get_band(self):
        thresh = .3
        return (self.value - thresh, self.value + thresh)

    def get_firms(self):
        apps = self.apps_per_day
        thresh = .2
        a,b = get_clips(self.value, thresh, 0, 1.)
        draws = stats.truncnorm.rvs(a,b, size = apps)
        draws = [d*thresh+self.value for d in draws]
        firms = [math.floor(d*self.num_firms) for d in draws]
        return firms

    def send_application(self):
        if self['labor'] > 0:
            firms = self.get_firms()
            for firm in firms:
                self.sell(('firm', firm), 
                      good = 'application', 
                      quantity = 1, 
                      price = 1/self.value)
            self.apps_sent += len(firms)
    
    def review_offers(self):
        offers = self.get_offers('labor', descending = True)

        # if no offers, do nothing
        if len(offers) == 0:
            return

        # count offers
        self.offers_received += len(offers)
        best_offer, *others = offers

        self.accept(best_offer)
        for o in others:
            self.reject(o)


    def get_value(self):
        return self.value

    def get_history(self):
        return self.apps_sent, self.offers_received
    
class Firm(Agent):
    def init(self, parameters):
        val = 1 / parameters['num_firms']
        moneys = val * (self.id - 1)
        moneys = moneys if moneys > 0 else val
        self.create('money', moneys)

    def send_offers(self):
        poss = self.possessions()
        applicants = self.get_offers('application')
        if not poss.get('labor'):
            if applicants:
                best_applicant, *others = applicants
                if 1/best_applicant.price >= self['money']:
                    self.buy(best_applicant.sender, 
                             good = 'labor', 
                             quantity = 1, 
                             price = self['money'])
```

```
def simulate(ps, name = 'basic'):
    sim = Simulation(name=name, processes = 1)
    firms = sim.build_agents(Firm, 'firm', number = ps['num_firms'], parameters = ps)
    workers = sim.build_agents(Worker, 'worker', number = ps['num_workers'],  parameters = ps)
    u = ps['num_workers']
    diff = 1000
    i = 0 
    while diff > 20:
    # for i in range(1):
        sim.advance_round(i)
        workers.send_application()
        firms.send_offers()
        workers.review_offers()
        # calculate unemployment
        new_u = np.array([w.get('labor') for w in workers.possessions()]).sum()
        diff = u - new_u
        u = new_u
        i += 1
        print(u)

    return workers, firms
```

```
def get_wages(workers):
    return np.array([w.get('money') for w in workers.possessions()])

def get_value(workers):
    return np.array([v for v in workers.get_value()])

def plot_xy(workers):
    y = get_wages(workers)
    X = get_value(workers)
    sns.regplot(x = X, y = y, fit_reg=False).set(xlim = (0,1.))

def plot_wage_dist(workers):
    wages = [i for i in get_wages(workers) if i]
    wages = [0 if i is None else i for i in wages]
    wages = sorted(wages)
    sns.regplot(x = range(len(workers)), y = wages, fit_reg=False)
```

```
%%capture

workers_3, firms = simulate({'num_firms': 3000, 'num_workers': 3000, 'apps_per_day': 3}) 
```

```
import pandas as pd

def get_results(workers):
    df = pd.DataFrame([h for h in workers.get_history()], 
                      columns = ['apps', 'offers'])
    df = df.assign(ratio = df.offers/df.apps, 
                   wage = get_wages(workers), 
                   value = get_value(workers))
    return df
```

```
value = .6
# low, hi = .995, .999
low, hi = value - .025, value + .025

df = get_results(workers_3)
df = df[(df.value > low) & (df.value < hi)]
w3_mean = df.ratio.mean()

df = get_results(workers_15)
df = df[(df.value > low) & (df.value < hi)]
w15_mean = df.ratio.mean()
```
